```python
#レーザ描写
fig=plt.figure(figsize=(10, 10))
plt.subplot(1,1,1)

X=[]
Y=[]

K=0.25
ini_deg = -45
draw_range = 5

for i in range(len(Data1.iloc[0][9:])):
#for i in range(2):
    if Data1.iloc[0][i+9] > draw_range:
        x = Data1.iloc[0][i+9] * np.cos(np.radians(i*K+ini_deg))
        y = Data1.iloc[0][i+9] * np.sin(np.radians(i*K+ini_deg))
        X.append(x)
        Y.append(y)
plt.scatter(X, Y,c="black",s=5)
    #print("i:",i , "   X: ", x , "    Y:  ",y)
    
```
